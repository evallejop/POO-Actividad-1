```python
class Edades:

    def calcular_edalberto(edjuan):
        return (2 * edjuan) / 3

    def calcular_edana(edjuan):
        return (4 * edjuan) / 3

    def calcular_edmama(edjuan, edalberto, edana):
        return edjuan + edalberto + edana

edjuan = 9
edalberto = Edades.calcular_edalberto(edjuan)
edana = Edades.calcular_edana(edjuan)
edmama = Edades.calcular_edmama(edjuan, edalberto, edana)

print(f"Las edades son: Alberto = {edalberto}, Juan = {edjuan}, Ana = {edana}, Mamá = {edmama}")
```
