```python
import math

class Calculos:

  @staticmethod
  def calcular_cuadrado(numero):
    return math.pow(numero, 2)

  @staticmethod
  def calcular_cubo(numero):
    return math.pow(numero,3)

numero = 4
calculo_cuadrado = Calculos.calcular_cuadrado(numero)
calculo_cubo = Calculos.calcular_cubo(numero)

print("El cuadrado del número es = " , calculo_cuadrado)
print("El cubo del número es = ", calculo_cubo)
```
